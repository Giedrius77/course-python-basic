### grupavimas

transponuojam lentelę, kad būtų patogiau dirbti su jos duomenimis

```python
visi_miestai = su_utena.transpose()
visi_miestai
```

pridedam regionus

```python
regionai = pd.Series([
    'Miestas', 
    'Miestas', 
    'Mazoji Lietuva',
    'Aukstaitija', 
    'Dzukija', 
    'Dzukija', 
    'Mazoji Lietuva', 
    'Aukstaitija'
])
laikini = visi_miestai.reset_index()
laikini['regionai'] = regionai
visi_miestai = laikini.set_index('index')
visi_miestai
```

sugrupuojam miestus pagal regionus, išvedam vidurkių lentelę

```python
regionais = visi_miestai.groupby('regionai')
regionais.mean(numeric_only=True)
```

regionų suma

```python
regionais.sum(numeric_only=True)
```

regionų skaičius (realiai visų duomenų bus vienodas)

```python
regionais.count()
```

min/max

```python
regionais.min()
```

```python
regionais.max() - regionais.min()
```

statistinė analizė vieno stulpelio duomens

```python
regionais[2024].describe()
```

```python
regionais[2024].describe().transpose()['Miestas']
```

### DF jungimas

pasidalinam duomenis

```python
iki_2010 = visi_miestai[['regionai', 2000, 2004, 2008]].copy()
po_2010 = visi_miestai[['regionai', 2012, 2016, 2020, 2024]].copy()
```

```python
top4 = visi_miestai[:'Panevezys'].copy()
bottom4 = visi_miestai['Alytus':].copy()
```

```python
pd.concat([top4, bottom4])
```

```python
pd.concat([top4, bottom4], axis=1)
```

```python
right4 = bottom4.reset_index()
right4.drop(columns='regionai', inplace=True)
right4
```

```python
left4 = top4.reset_index().drop(columns='regionai')
left4
```

```python
pd.concat([left4, right4], axis=1)
```

```python
pd.concat([iki_2010, po_2010], axis=1).drop(columns='regionai', axis=1)
```

```python
pd.merge(iki_2010, po_2010, on=['index', 'regionai'])
```

```python
iki_2010.join(po_2010, lsuffix='', rsuffix=':R').drop(columns='regionai:R')
```
