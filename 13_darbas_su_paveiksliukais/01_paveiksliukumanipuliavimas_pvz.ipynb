```python
# mano pavyzdiai su mano_face

from PIL import Image

foto = Image.open('mano_face.jpg') # priskiriame nauja kintamaji mano_face failui ir jo KELIA!!!!
foto # atidarys mano_face faila, bet reikia nuimti foto.size
foto.size # atidarys foto dydi
```

```python
# pakeisti foto dydi i koki nori
naujas_foto_dydis = (300, 400)
pakeistas_dysis = foto.resize(naujas_foto_dydis)
pakeistas_dysis.save('mano_face_mazas.jpg')
pakeistas_dysis
```

```python
# nusipiesti kvadrata ir taip issiaiskinti kur padaryti kirpima foto
from PIL import ImageDraw

def draw_square(pakeistas_dysis:Image, top_x, top_y, bottom_x, bottom_y, fill="red", width=1):
    draft = pakeistas_dysis.copy()
    draw = ImageDraw.Draw(draft)
    top_line = draw.line((top_x, top_y, bottom_x, top_y), fill=fill, width=width)
    left_line = draw.line((top_x, top_y, top_x, bottom_y), fill=fill, width=width)
    bottom_line = draw.line((top_x, bottom_y, bottom_x, bottom_y), fill=fill, width=width)
    right_line = draw.line((bottom_x, top_y, bottom_x, bottom_y), fill=fill, width=width)
    return draft

squared = draw_square(pakeistas_dysis, 50, 200, 120, 160, width=3, fill="#ffcc00")
squared.save('mano_face_kaire_akis.jpg')
squared
```
